# Step 5: Data Preprocessing

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Data Cleaning

# Handling missing values in 'Cuisines' column
imputer = SimpleImputer(strategy='most_frequent')
merged_df['Cuisines'] = imputer.fit_transform(merged_df[['Cuisines']])

# Encoding categorical variables using Label Encoding
label_encoder = LabelEncoder()
merged_df['Has Table booking'] = label_encoder.fit_transform(merged_df['Has Table booking'])
merged_df['Has Online delivery'] = label_encoder.fit_transform(merged_df['Has Online delivery'])
merged_df['Is delivering now'] = label_encoder.fit_transform(merged_df['Is delivering now'])
merged_df['Switch to order menu'] = label_encoder.fit_transform(merged_df['Switch to order menu'])
merged_df['Rating color'] = label_encoder.fit_transform(merged_df['Rating color'])
merged_df['Rating text'] = label_encoder.fit_transform(merged_df['Rating text'])

#Feature Scaling

# Scaling numerical features using StandardScaler
numerical_columns = ['Longitude', 'Latitude', 'Average Cost for two', 'Votes']
scaler = StandardScaler()
merged_df[numerical_columns] = scaler.fit_transform(merged_df[numerical_columns])

# Displaying the first few rows of the preprocessed dataframe
print("Preprocessed Dataset:")
print(merged_df.head())

# One-hot encode the 'Cuisines' column
cuisines_dummies = merged_df['Cuisines'].str.get_dummies(sep=', ')
merged_df = pd.concat([merged_df, cuisines_dummies], axis=1)

# Dropping the original 'Cuisines' column and other unnecessary columns
merged_df.drop(['Cuisines', 'Restaurant Name', 'Address', 'Locality', 'Locality Verbose', 'Currency', 'Rating color', 'Rating text', 'Country'], axis=1, inplace=True)

# Displaying the first few rows of the modified dataframe
print("Modified Dataset:")
print(merged_df.head())
```

# Step 7: Feature Selection and Train-Test Split

```python
from sklearn.model_selection import train_test_split

# Selecting relevant features for prediction
selected_features = ['Longitude', 'Latitude', 'Has Table booking', 'Has Online delivery',
                     'Is delivering now', 'Switch to order menu', 'Price range', 'Aggregate rating', 'Votes']

# Extracting the selected features and target variables
X = merged_df[selected_features]
y_average_cost = merged_df['Average Cost for two']
y_price_range = merged_df['Price range']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train_average_cost, y_test_average_cost, y_train_price_range, y_test_price_range = train_test_split(
    X, y_average_cost, y_price_range, test_size=0.2, random_state=42)

# Displaying the shapes of the training and testing sets
print("Shapes of Training and Testing Sets:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train_average_cost:", y_train_average_cost.shape)
print("y_test_average_cost:", y_test_average_cost.shape)
print("y_train_price_range:", y_train_price_range.shape)
print("y_test_price_range:", y_test_price_range.shape)
```
